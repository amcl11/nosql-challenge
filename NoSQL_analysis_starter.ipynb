### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_query = {
    "LocalAuthorityName": {"$regex": "London"},
    "RatingValue": {"$gte": 4}
          }

# Use count_documents to display the number of documents in the result
establishments.count_documents(london_query)
```

```python
# Display the first document in the results using pprint
first_london_doc = establishments.find_one(london_query)

pprint(first_london_doc)
```

```python
# Convert the result to a Pandas DataFrame
london_results = establishments.find(london_query)
london_df = pd.DataFrame(london_results)
# Display the number of rows in the DataFrame
london_rows = len(london_df)
print(london_rows)

```

```python
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.083840

# Calculate longitude range
longitude_range = (longitude - degree_search, longitude + degree_search)

# Calculate latitude range
latitude_range = (latitude - degree_search, latitude + degree_search)

rating_query = {
    "RatingValue": 5,
    "geocode.longitude": {"$gte": longitude_range[0], "$lte": longitude_range[1]},
    "geocode.latitude": {"$gte": latitude_range[0], "$lte": latitude_range[1]}
}
sort = [("scores.Hygiene", 1)]

limit = 5

top_5_results = list(establishments.find(rating_query).sort(sort).limit(limit))

# Print the results
pprint(top_5_results)
```

```python
# Convert result to Pandas DataFrame
rating_df = pd.DataFrame(top_5_results)
```
